```python
class BinaryAccuracy:
    def __init__(
            self,
            logits: bool = True,
            reduction: typing.Callable[
                [
                    torch.Tensor,
                ],
                torch.Tensor,
            ] = torch.mean,
    ):
        self.logits = logits
        if logits:
            self.threshold = 0
        else:
            self.threshold = 0.5

        self.reduction = reduction

    def __call__(self, y_pred, y_true):
        return self.reduction(((y_pred > self.threshold) == y_true.bool()).float())
```

```python
import random

for root, folders, files in os.walk("data/dice/0/"):
    pictures = [path.join(root, file) for file in files]

choices = random.choices(pictures, k=50)

for root, folders, files in os.walk("data/defect/anomalous_dice/"):
    pictures = [path.join(root, file) for file in files]
    
choices.extend(random.choices(pictures, k=15))

random.shuffle(choices)

choices
```

```python
out = []
stats = (0.5, 0.5, 0.5), (0.5, 0.5, 0.5)
tt_transform = tt.Compose([tt.Resize(128),
                             tt.CenterCrop(128),
                             tt.ToTensor(),
                             tt.Normalize(*stats)])
threshold = 0.7
disc = torch.load("./discriminator.pt", map_location=torch.device('cpu'))
for img_path in choices:
    img = Image.open(img_path).convert('RGB')
    img_tensor = tt_transform(img).unsqueeze_(0)
    score = disc(img_tensor)[0].item()
    score = round(score, 2)
    predict = 1 if score >= threshold else 0
    if "anomalous_dice" in img_path:
        out.append([img_path, score, 0, predict])
    else:
        out.append([img_path, score, 1, predict])

df = pd.DataFrame(out, columns=["path", "score", "real_class", "predicted_class"])
```
